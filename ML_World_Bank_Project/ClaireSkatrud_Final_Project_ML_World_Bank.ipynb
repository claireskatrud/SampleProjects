## Address outliers

```python
numeric_cols = working_data_2018.select_dtypes(include=[np.number])

outlier_summary = {}

for col in numeric_cols.columns:
    s = numeric_cols[col].dropna()
    
    mean = s.mean()
    std = s.std()
    
    lower = mean - 3 * std
    upper = mean + 3 * std
    
    #find outlier indices
    outliers = working_data_2018[(working_data_2018[col] < lower) | 
                                 (working_data_2018[col] > upper)][col]
    
    outlier_summary[col] = {
        'num_outliers': len(outliers),
        'percentage_outliers': len(outliers) / len(s) * 100,
        'outlier_indices': outliers.index.tolist(),
        'lower_bound': lower,
        'upper_bound': upper
    }

outlier_df = pd.DataFrame(outlier_summary).T
outlier_df = outlier_df.sort_values('num_outliers', ascending=False)

outlier_df[outlier_df['num_outliers']>0]
```

### Drop nulls, impute, winsorize

```python
from sklearn.preprocessing import RobustScaler

working_data = working_data_2018.copy()

#remove rows with too many nulls
working_data['num_nulls'] = working_data.isna().sum(axis=1)
working_data = working_data[working_data['num_nulls'] <= 20]
working_data = working_data.drop('num_nulls', axis=1)


numeric_cols = working_data.select_dtypes(include='number').columns.tolist()
feature_numeric_cols = [c for c in numeric_cols if c != target_col]

#impute missing values before scaling
indicator_cols = feature_numeric_cols

for col in indicator_cols:
    working_data[col] = working_data.groupby('Region')[col].transform(
        lambda x: x.fillna(x.median())
    )
    working_data[col] = working_data.groupby('Income Group')[col].transform(
        lambda x: x.fillna(x.median())
    )
    working_data[col] = working_data[col].fillna(working_data[col].median())

#winsorize skewed features before scaling
skewed_features = working_data[feature_numeric_cols].skew().abs()
skewed_features = skewed_features[skewed_features > 1].index.tolist()

print(f"Heavily skewed features ({len(skewed_features)}): {skewed_features[:10]} ...")

def winsorize_series(s, lower=0.01, upper=0.99):
    """Clip values at the lower and upper percentiles."""
    lower_val = s.quantile(lower)
    upper_val = s.quantile(upper)
    return s.clip(lower=lower_val, upper=upper_val)

for col in skewed_features:
    working_data[col] = winsorize_series(working_data[col])

X = working_data.drop([target_col, 'Country Name', 'Country Code'], axis=1)
X = pd.get_dummies(X, columns=['Region', 'Income Group'], drop_first=True)
y = working_data[target_col]
```

### Adress null values and do hyper parameter tuning (random forest)

```python
from sklearn.preprocessing import RobustScaler
from sklearn.model_selection import KFold
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import cross_val_score
```

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import RobustScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score

numeric_feature_cols_in_X = [c for c in X.columns if c in feature_numeric_cols]
dummy_cols = [c for c in X.columns if c not in feature_numeric_cols]

print(f"Numeric features to scale: {len(numeric_feature_cols_in_X)}")
print(f"Dummy variables (not scaled): {len(dummy_cols)}")


preprocessor = ColumnTransformer(
    transformers=[
        ('num', RobustScaler(), numeric_feature_cols_in_X),
        ('cat', 'passthrough', dummy_cols)
    ],
    remainder='passthrough' 
)

#scaling happens within each fold
rf_pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('rf', RandomForestRegressor(
        n_estimators=500,
        max_depth=None,
        random_state=42,
        n_jobs=-1
    ))
])

#cross validation
cv_r2_scores = cross_val_score(rf_pipeline, X, y, cv=5, scoring='r2')
print("Cross-validated R² mean:", np.mean(cv_r2_scores))
print("Cross-validated R² std:", np.std(cv_r2_scores))

scores = cross_val_score(rf_pipeline, X, y, cv=5, scoring='neg_mean_squared_error')
rmse_scores = np.sqrt(-scores)
print("RF RMSE mean:", np.mean(rmse_scores))
print("RF RMSE std:", np.std(rmse_scores))
```

```python
from sklearn.model_selection import KFold
from sklearn.preprocessing import RobustScaler

kf = KFold(n_splits=5, shuffle=True, random_state=42)
importances_list = []

numeric_feature_cols_in_X = [c for c in X.columns if c in feature_numeric_cols]

for train_index, test_index in kf.split(X):
    X_train_fold = X.iloc[train_index]
    X_test_fold = X.iloc[test_index]
    y_train = y.iloc[train_index]
    
    scaler = RobustScaler()
    X_train_scaled = X_train_fold.copy()
    
    X_train_scaled[numeric_feature_cols_in_X] = scaler.fit_transform(X_train_fold[numeric_feature_cols_in_X])
    
    # Fit Random Forest on scaled training data
    rf = RandomForestRegressor(
        n_estimators=500,
        max_depth=None,
        random_state=42,
        n_jobs=-1
    )
    rf.fit(X_train_scaled, y_train)
    importances_list.append(rf.feature_importances_)

mean_importances = np.mean(importances_list, axis=0)
importances = pd.Series(mean_importances, index=X.columns).sort_values(ascending=False)
print("\nTop 20 Most Important Features:")
print(importances.head(20))
```

Hyperparameter tuning on all features

```python
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import RobustScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_error

numeric_feature_cols_in_X = [c for c in X.columns if c in feature_numeric_cols]
dummy_cols = [c for c in X.columns if c not in feature_numeric_cols]

preprocessor = ColumnTransformer(
    transformers=[
        ('num', RobustScaler(), numeric_feature_cols_in_X),
        ('cat', 'passthrough', dummy_cols)
    ],
    remainder='passthrough'
)

rf_pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('rf', RandomForestRegressor(random_state=42))
])

param_grid = {
    'rf__n_estimators': [100, 250, 500],
    'rf__max_depth': [None, 5, 10]
}

# GridSearchCV with pipeline
grid_search = GridSearchCV(
    rf_pipeline, 
    param_grid, 
    cv=5, 
    scoring='neg_mean_squared_error', 
    n_jobs=-1,
    verbose=1
)
grid_search.fit(X, y)

best_params = grid_search.best_params_
best_rmse = np.sqrt(-grid_search.best_score_)

print("Best Parameters:", best_params)
print("Best Score (RMSE):", best_rmse)

#get best pipeline (includes preprocessor + model)
best_pipeline = grid_search.best_estimator_

#do final cross-validation with best pipeline
final_rmse_scores = np.sqrt(
    -cross_val_score(best_pipeline, X, y, cv=5, scoring='neg_mean_squared_error')
)

final_cv_r2_scores = cross_val_score(best_pipeline, X, y, cv=5, scoring='r2')

print("\nFinal R² mean:", np.mean(final_cv_r2_scores))
print("Final R² std:", np.std(final_cv_r2_scores))
print("Final RMSE mean:", np.mean(final_rmse_scores))
print("Final RMSE std:", np.std(final_rmse_scores))
```

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from sklearn.model_selection import KFold, cross_val_score
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import RobustScaler
from scikeras.wrappers import KerasRegressor # Use KerasRegressor from scikeras
from tensorflow.keras import layers
from tensorflow.keras import regularizers
from tensorflow.keras.layers import Input
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('num', RobustScaler(), numeric_feature_cols_in_X),
        ('cat', 'passthrough', dummy_cols)
    ],
    remainder='passthrough'
)
INPUT_SHAPE = X.shape[1]
```

```python
def build_baseline_model(input_shape=INPUT_SHAPE):
    model = keras.Sequential([
        # Only one hidden layer
        Input(shape=(input_shape,)),
        layers.Dense(32, activation='relu'),
        # Output layer for regression (linear activation)
        layers.Dense(1, activation='linear')
    ])
    model.compile(optimizer='adam', loss='mse', metrics=['mae'])
    return model
```

```python
def build_large_model(input_shape=INPUT_SHAPE):
    model = keras.Sequential([
        Input(shape=(input_shape,)),
        layers.Dense(256, activation='relu'),
        layers.Dense(128, activation='relu'),
        layers.Dense(64, activation='relu'),
        layers.Dense(1, activation='linear')
    ])
    model.compile(optimizer='adam', loss='mse', metrics=['mae'])
    return model
```

```python
def build_regularized_model(input_shape=INPUT_SHAPE):
    model = keras.Sequential([
        Input(shape=(input_shape,)),
        #L2 regularization on the kernel weights (Lambda=0.001)
        layers.Dense(256, activation='relu', kernel_regularizer=regularizers.l2(0.001)),
        
        #dropout to randomly switch off 50% of neurons
        layers.Dropout(0.5),
        
        layers.Dense(128, activation='relu', kernel_regularizer=regularizers.l2(0.001)),
        layers.Dropout(0.5),
        
        layers.Dense(64, activation='relu'),
        layers.Dense(1, activation='linear')
    ])
    model.compile(optimizer='adam', loss='mse', metrics=['mae'])
    return model
```

```python
#hyperparameters for KerasRegressor (same for all models)
NN_FIT_PARAMS = {
    'epochs': 100, 
    'batch_size': 32, 
    'verbose': 0
}

#Baseline Model CV
keras_baseline = KerasRegressor(
    model=build_baseline_model, 
    **NN_FIT_PARAMS
)
pipeline_baseline = Pipeline([
    ('preprocessor', preprocessor),
    ('nn', keras_baseline)
])

#RMSE Scoring for the full pipeline (returns negative RMSE)
neg_rmse_baseline = cross_val_score(
    pipeline_baseline, 
    X, 
    y, 
    cv=5, 
    scoring='neg_root_mean_squared_error'
)
#Convert the negative score to positive RMSE
rmse_baseline = -neg_rmse_baseline

print("## Baseline Model Results")
print(f"Mean CV RMSE: {np.mean(rmse_baseline):.4f}")
print("---")

#Large Overfitting Model CV
keras_large = KerasRegressor(
    model=build_large_model, 
    **NN_FIT_PARAMS
)
pipeline_large = Pipeline([
    ('preprocessor', preprocessor),
    ('nn', keras_large)
])

neg_rmse_large = cross_val_score(
    pipeline_large, 
    X, 
    y, 
    cv=5, 
    scoring='neg_root_mean_squared_error'
)
rmse_large = -neg_rmse_large

print("## Large Model Results (Likely Overfitting)")
print(f"Mean CV RMSE: {np.mean(rmse_large):.4f}")
print("---")

# Regularized Model CV
keras_regularized = KerasRegressor(
    model=build_regularized_model, 
    **NN_FIT_PARAMS
)
pipeline_regularized = Pipeline([
    ('preprocessor', preprocessor),
    ('nn', keras_regularized)
])

neg_rmse_regularized = cross_val_score(
    pipeline_regularized, 
    X, 
    y, 
    cv=5, 
    scoring='neg_root_mean_squared_error'
)
rmse_regularized = -neg_rmse_regularized

print("## Regularized Model Results")
print(f"Mean CV RMSE: {np.mean(rmse_regularized):.4f}")
print("---")
```

#### Hyperparameter tuning for regularization parameters

```python
def build_regularized_model(input_shape=INPUT_SHAPE, l2_lambda=0.001, dropout_rate=0.5):
    """
    Builds a Keras Sequential model with customizable L2 and Dropout.
    These parameters are passed directly from GridSearchCV.
    """
    tf.keras.backend.clear_session() #clear residual graphs 

    model = keras.Sequential([
        Input(shape=(input_shape,)),
        

        layers.Dense(256, activation='relu', kernel_regularizer=regularizers.l2(l2_lambda)),
        layers.Dropout(dropout_rate),
    
        layers.Dense(128, activation='relu', kernel_regularizer=regularizers.l2(l2_lambda)),
        layers.Dropout(dropout_rate),
        layers.Dense(1, activation='linear')
    ])
    
    
    model.compile(optimizer='adam', loss='mse', metrics=['mae'])
    return model
```

```python
NN_FIT_PARAMS = {
    'epochs': 150,
    'batch_size': 8, 
    'verbose': 0
}

keras_regressor = KerasRegressor(
    model=build_regularized_model, 
    **NN_FIT_PARAMS
)

pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('nn', keras_regressor) 
])


param_grid = {
    'nn__model__l2_lambda': [1e-4, 1e-3, 5e-3, 1e-2], 
    'nn__model__dropout_rate': [0.2, 0.4, 0.5, 0.6], 
}

scoring_metric = 'neg_root_mean_squared_error'

print(f"Starting GridSearchCV with {len(param_grid['nn__model__l2_lambda']) * len(param_grid['nn__model__dropout_rate'])} total fits (5 folds each). This may take some time...")

grid_search = GridSearchCV(
    estimator=pipeline,
    param_grid=param_grid,
    scoring=scoring_metric, 
    cv=KFold(n_splits=5, shuffle=True, random_state=42),
    verbose=2,
    n_jobs=-1 
)

grid_search.fit(X, y.to_numpy())


print("\n\n--------------------------------------------")
print("           GRID SEARCH RESULTS")
print("--------------------------------------------")
print("Best Parameters Found:")

print(grid_search.best_params_)


best_rmse = -grid_search.best_score_
print(f"\nBest Cross-Validated RMSE: {best_rmse:.4f}")
print("--------------------------------------------")


best_pipeline = grid_search.best_estimator_
```

```python
L2_OPTIMAL = 0.001
DROPOUT_OPTIMAL = 0.4

NN_FIT_PARAMS = {
    'epochs': 150,
    'batch_size': 8, 
    'verbose': 0
}



def build_original_model(input_shape=INPUT_SHAPE, l2=L2_OPTIMAL, dropout=DROPOUT_OPTIMAL):
    """Architecture from your previous Regularized Model: 256 -> 128 -> 1."""
    tf.keras.backend.clear_session()
    model = keras.Sequential([
        Input(shape=(input_shape,)), 
        
        layers.Dense(256, activation='relu', kernel_regularizer=regularizers.l2(l2)),
        layers.Dropout(dropout),
        
        layers.Dense(128, activation='relu', kernel_regularizer=regularizers.l2(l2)),
        layers.Dropout(dropout),
        
        layers.Dense(1, activation='linear')
    ])
    model.compile(optimizer='adam', loss='mse', metrics=['mae'])
    return model

def build_wide_shallow_model(input_shape=INPUT_SHAPE, l2=L2_OPTIMAL, dropout=DROPOUT_OPTIMAL):
    """Wide and Shallow Architecture: 512 -> 1."""
    tf.keras.backend.clear_session()
    model = keras.Sequential([
        Input(shape=(input_shape,)),
        
        layers.Dense(512, activation='relu', kernel_regularizer=regularizers.l2(l2)), # Wider first layer
        layers.Dropout(dropout),
        
        layers.Dense(1, activation='linear')
    ])
    model.compile(optimizer='adam', loss='mse', metrics=['mae'])
    return model

def build_deep_narrow_model(input_shape=INPUT_SHAPE, l2=L2_OPTIMAL, dropout=DROPOUT_OPTIMAL):
    """Deep and Narrow Architecture: 128 -> 64 -> 32 -> 1."""
    tf.keras.backend.clear_session()
    model = keras.Sequential([
        Input(shape=(input_shape,)),
        
        layers.Dense(128, activation='relu', kernel_regularizer=regularizers.l2(l2)),
        layers.Dropout(dropout),
        
        layers.Dense(64, activation='relu', kernel_regularizer=regularizers.l2(l2)),
        layers.Dropout(dropout),
        
        layers.Dense(32, activation='relu', kernel_regularizer=regularizers.l2(l2)), # Added another narrow layer
        layers.Dropout(dropout),
        
        layers.Dense(1, activation='linear')
    ])
    model.compile(optimizer='adam', loss='mse', metrics=['mae'])
    return model



def evaluate_architecture(build_fn, name):
    print(f"-> Starting evaluation for: {name}...")
    keras_regressor = KerasRegressor(model=build_fn, **NN_FIT_PARAMS)
    
    pipeline = Pipeline([
        ('preprocessor', preprocessor),
        ('nn', keras_regressor)
    ])
    

    neg_rmse_scores = cross_val_score(
        pipeline, 
        X, y.to_numpy(),
        cv=KFold(n_splits=5, shuffle=True, random_state=42),
        scoring='neg_root_mean_squared_error',
        n_jobs=-1
    )
    rmse_scores = -neg_rmse_scores
    print(f"-> {name} complete.")
    return rmse_scores


rmse_original = evaluate_architecture(build_original_model, "Original (256-128)")

rmse_wide_shallow = evaluate_architecture(build_wide_shallow_model, "Wide/Shallow (512-1)")

rmse_deep_narrow = evaluate_architecture(build_deep_narrow_model, "Deep/Narrow (128-64-32)")



print("\n\n--------------------------------------------")
print("          ARCHITECTURE COMPARISON")
print("--------------------------------------------")
print("All models use L2=0.001 and Dropout=0.4.")
print(f"1. Original (256 -> 128) Mean CV RMSE: {np.mean(rmse_original):.4f}")
print(f"2. Wide/Shallow (512 -> 1) Mean CV RMSE: {np.mean(rmse_wide_shallow):.4f}")
print(f"3. Deep/Narrow (128 -> 64 -> 32) Mean CV RMSE: {np.mean(rmse_deep_narrow):.4f}")
print("--------------------------------------------")
```
